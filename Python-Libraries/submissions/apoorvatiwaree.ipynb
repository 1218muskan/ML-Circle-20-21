#### 1. Write a NumPy program to create an array of 10 zeros,10 ones, 10 fives. Write a NumPy program to create an array of the integers from 30 to70

```python
import numpy as np
zeros=np.zeros(10)
ones=np.ones(10)
fives=np.full(10,5)
print("Array of 0s =",zeros,"\nArray of 1s =",ones,"\nArray of 5s=",fives)

integers=np.arange(30,71)
print("\nArray of integers from 30 to 70 =",integers)
```

#### 3.Write a NumPy program to create a vector with values from 0 to 20 and change the sign of the numbers in the range from 9 to 15.

```python
b=np.arange(0,21)
print("\nOriginal Array",b)
b[9:16]=-1*b[9:16]
print("\nFinal Array=",b)
```

#### 4.Write a NumPy program to compute the x and y coordinates for points on a sine curve and plot the points using matplotlib.

```python
import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(0, 3 * np.pi, 0.1) 
y = np.sin(x)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Sine Function Curve')
plt.plot(x, y)
plt.show()
```

#### 5.Write a NumPy program to create a random 10x4 array and extract the first five rows of the array and store them into a variable. (hint: using numpy random module)

```python
m= np.random.rand(10, 4)
print("\nOriginal Array =",m)
m_new= m[:5, :]
print("\nFirst 5 rows=",m_new)
```
